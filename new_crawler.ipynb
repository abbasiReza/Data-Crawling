```python
import os
files=os.listdir('new_combs/')
```

```python
remove_index=[]
for file in files:
 
    with open('new_combs/'+file,'r')as f:
        lines=f.readlines()
    for i in range(len(lines)):
        lines[i]=lines[i].replace('\n','')
    for a in lines:
        if int(a) not in remove_index:
            remove_index.append(int(a))
```
